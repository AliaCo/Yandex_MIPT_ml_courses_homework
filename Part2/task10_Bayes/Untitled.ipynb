```python
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection, naive_bayes
```

```python
X_bc, y_bc = datasets.load_breast_cancer(return_X_y=True, as_frame=True)
X_d, y_d = datasets.load_digits(return_X_y=True, as_frame=True)
```

```python
res_B = model_selection.cross_val_score(naive_bayes.BernoulliNB(), X_bc, y_bc)
print('BernoulliNB ', np.mean(res_B))

res_M = model_selection.cross_val_score(naive_bayes.MultinomialNB(), X_bc, y_bc)
print('MultinomialNB ', np.mean(res_M))

res_G = model_selection.cross_val_score(naive_bayes.GaussianNB(), X_bc, y_bc)
print('GaussianNB ', np.mean(res_G))
```

```python
res_B = model_selection.cross_val_score(naive_bayes.BernoulliNB(), X_d, y_d)
print('BernoulliNB ', np.mean(res_B))

res_M = model_selection.cross_val_score(naive_bayes.MultinomialNB(), X_d, y_d)
print('MultinomialNB ', np.mean(res_M))

res_G = model_selection.cross_val_score(naive_bayes.GaussianNB(), X_d, y_d)
print('GaussianNB ', np.mean(res_G))
```
